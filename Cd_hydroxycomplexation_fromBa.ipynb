# Cd hydroxycomplexation

The overall complexation constants for Cd$^{2+}$ with OH$^-$ are $\log{\beta_{f1}}=4.1$, $\log{\beta_{f2}}=7.7$, $\log{\beta_{f3}}=10.3$ and $\log{\beta_{f4}}=12$ corresponding to the following reactions (in order)

$\mathrm{Cd^{2+}}+\mathrm{OH^-} \leftrightharpoons \mathrm{CdOH^+}$

$\mathrm{Cd^{2+}}+2\mathrm{OH^-} \leftrightharpoons \mathrm{Cd(OH)_2^0}$

$\mathrm{Cd^{2+}}+3\mathrm{OH^-} \leftrightharpoons \mathrm{Cd(OH)_3^-}$

$\mathrm{Cd^{2+}}+4\mathrm{OH^-} \leftrightharpoons \mathrm{Cd(OH)_4^{2-}}$

We derived the alpha expressions for these 5 species in class

$\alpha_{\mathrm{Cd^{2+}}} = \dfrac{[\mathrm{H^{+}}]^4}{\beta_{a4}+\beta_{a3}[\mathrm{H^+}]+\beta_{a2}[\mathrm{H^+}]^2+\beta_{a1}[\mathrm{H^+}]^3+[\mathrm{H^{+}}]^4}$


$\alpha_{\mathrm{CdOH^{+}}} = \dfrac{\beta_{a1}[\mathrm{H^+}]^3}{\beta_{a4}+\beta_{a3}[\mathrm{H^+}]+\beta_{a2}[\mathrm{H^+}]^2+\beta_{a1}[\mathrm{H^+}]^3+[\mathrm{H^{+}}]^4}$


$\alpha_{\mathrm{Cd(OH)_2^{0}}} = \dfrac{\beta_{a2}[\mathrm{H^+}]^2}{\beta_{a4}+\beta_{a3}[\mathrm{H^+}]+\beta_{a2}[\mathrm{H^+}]^2+\beta_{a1}[\mathrm{H^+}]^3+[\mathrm{H^{+}}]^4}$


$\alpha_{\mathrm{Cd(OH)_3^{-}}} = \dfrac{\beta_{a3}[\mathrm{H^+}]}{\beta_{a4}+\beta_{a3}[\mathrm{H^+}]+\beta_{a2}[\mathrm{H^+}]^2+\beta_{a1}[\mathrm{H^+}]^3+[\mathrm{H^{+}}]^4}$

$\alpha_{\mathrm{Cd(OH)_4^{2-}}} = \dfrac{\beta_{a4}}{\beta_{a4}+\beta_{a3}[\mathrm{H^+}]+\beta_{a2}[\mathrm{H^+}]^2+\beta_{a1}[\mathrm{H^+}]^3+[\mathrm{H^{+}}]^4}$

In class we learned how to quickly sketch the distribution diagram using the step-wise formation constants.  We can make a more quantitative sketch though using these expressions (note, since pH would be a known quantity we do not need to make any additional assumptions.).

```octave
%plot -s 600,500 -f 'svg'
logB1=4.1; logB2=7.7; logB3=10.3; logB4=12; Kw=1e-14; logKw=log10(Kw);
logBa1=logB1+logKw; logBa2=logB2+2*logKw; logBa3=logB3+3*logKw; logBa4=logB4+4*logKw;
Ba1=10^logBa1; Ba2=10^logBa2; Ba3=10^logBa3; Ba4=10^logBa4;

pH=6:0.1:14; H=10.^(-pH); OH=Kw./H;
denominator=H.^4+Ba1*H.^3+Ba2*H.^2+Ba3*H+Ba4;
alphaCd=(H.^4)./denominator;
alphaCdOH=(H.^3*Ba1)./denominator;
alphaCdOH2=(H.^2*Ba2)./denominator;
alphaCdOH3=(H*Ba3)./denominator;
alphaCdOH4=(Ba4)./denominator;

plot(pH,alphaCd,'b','linewidth',2)
hold on
plot(pH,alphaCdOH,'g','linewidth',2)
plot(pH,alphaCdOH2,'m','linewidth',2)
plot(pH,alphaCdOH3,'k','linewidth',2)
plot(pH,alphaCdOH4,'c','linewidth',2)
xlabel('pH'); ylabel('\alpha');
legend('Cd^{2+}','CdOH^+','Cd(OH)2','Cd(OH)3^-','Cd(OH)4^{2-}','location','west')
set(gca,'linewidth',2)
title('Distribution diagram, Cd hydroxo complexation')

% add stepwise values (pH=pKa but we need to convert because we were given overall association constants)

plot([9.9 9.9],[0 1],'k:','linewidth',2)
plot([10.4 10.4],[0 1],'k:','linewidth',2)
plot([11.4 11.4],[0 1],'k:','linewidth',2)
plot([12.3 12.3],[0 1],'k:','linewidth',2)

```
