# Diprotic acid (H2A)

There are two stepwise dissociation equilibria

$\mathrm{H_2A} = \mathrm{HA^-} + \mathrm{H^+}$

$K_{a1} = \dfrac{[\mathrm{H}^+][\mathrm{HA^-}]}{[\mathrm{H_2A}]}$

and

$\mathrm{HA^-} = \mathrm{A^{2-}} + \mathrm{H^+}$

$K_{a2} = \dfrac{[\mathrm{H}^+][\mathrm{A^{2-}}]}{[\mathrm{HA^-}]}$

We derived in class the three alpha relationships (remember $\alpha$ is the fraction each species contributes to the total)

$\alpha_{H2A}=\dfrac{[\mathrm{H}^+]^2}{[\mathrm{H}^+]^2+K_{a1}[\mathrm{H^+}]+K_{a1}K_{a2}}$

$\alpha_{HA-}=\dfrac{K_{a1}[\mathrm{H^+}]}{[\mathrm{H}^+]^2+K_{a1}[\mathrm{H^+}]+K_{a1}K_{a2}}$

$\alpha_{A2-}=\dfrac{K_{a1}K_{a2}}{[\mathrm{H}^+]^2+K_{a1}[\mathrm{H^+}]+K_{a1}K_{a2}}$

For some arbitrary diprotic weak acid we can plot the distribution (speciation) diagram.  Let's assume pKa1=4 and pKa2=8.


```octave
pKa1 = input("Enter pKa1 value: ");
pKa2 = input("Enter pKa2 value: ");
pKa1=6.3; pKa2=10.3;
Ka1=10^-pKa1; Ka2=10^-pKa2;
pH=2:0.1:12; H=10.^-pH;

denominator=H.^2+Ka1*H+Ka1*Ka2;
alphaH2A=(H.^2)./denominator;
alphaHA=(Ka1*H)./denominator;
alphaA=1-alphaH2A-alphaHA;

plot(pH,alphaH2A,'linewidth',2,pH,alphaHA,'linewidth',2,pH,alphaA,'linewidth',2)
set(gca,'fontsize',12,'linewidth',2)
legend('H_2A','HA^-','A^{2-}')
title('Distribution Diagram for H2A')
xlabel('pH'); ylabel('\alpha','fontsize',16)

hold on
plot([pKa1 pKa1],[0 1],'k:','linewidth',2)
plot([pKa2 pKa2],[0 1],'k:','linewidth',2)
plot([2 12],[0.5 0.5],'k:','linewidth',2)
```
